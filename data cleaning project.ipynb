```python
import pandas as pd 
df=pd.read_excel("Customer Call List.xlsx")
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
df=df.drop(columns="Not_Useful_Column")
```

```python
df
```

# to remove unwanted things betwee columns

```python
df["Last_Name"]=df["Last_Name"].str.strip("_")
df["Last_Name"]=df["Last_Name"].str.strip("...")
df["Last_Name"]=df["Last_Name"].str.strip("/")

''' or we can also also
df["Last_Name"]=df["Last_Name"].str.strip("123/_...")
this is used when we have to remove all unwanted thing from our column in one step'''
```

```python
df
```

# now we need to clean the phone no 

```python
df["Phone_Number"]=df["Phone_Number"].astype(str)
df["Phone_Number"]=df["Phone_Number"].replace(r'\D',"",regex=True)
```

```python
df
```

# now we have to put "-" betwee the phone number

```python
df
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda X : X[0:3]+"-"+ X[3:6]+"-"+ X[6:10])
```

```python
df["Phone_Number"]=df["Phone_Number"].replace(r'\D',"",regex=True)
```

```python
df
```

```python
df[["Street_Address","State","Zip_Code"]]=df["Address"].str.split(",",n=2,expand=True)
```
